```python
# 取得page_name欄位並去除重複值
page_name_list = df_other["page_name"].drop_duplicates()

# 重置索引並轉換為列表
page_name_list = page_name_list.reset_index(drop=True).tolist()
```

```python
page_name_list
```

```python
import re
import pandas as pd

# 初始化一個列表來存儲匹配到的名字
all_matched_names = []

# 迭代遍歷 df_main 中的每一行
for content in df_main['content']:
    matched_names = []
    for name in page_name_list:
        if re.search(re.escape(name), content):
            matched_names.append(name)
    # 將匹配到的名字用逗號分隔
    matched_names_str = ', '.join(matched_names)
    # 分割字符串並存儲在列表中
    matched_names_list = [x.strip() for x in matched_names_str.strip('[]').split(',') if x.strip()]
    all_matched_names.append(matched_names_list)

# 添加一個新列來存儲匹配到的名字
df_main['matched_names'] = all_matched_names

# 通過展開匹配名字的列表來創建一個新的 DataFrame
df_exploded = df_main.explode('matched_names')

# 刪除匹配名字為空的行
df_exploded = df_exploded.dropna(subset=['matched_names'])

# 重置索引
df_exploded.reset_index(drop=True, inplace=True)

df_exploded
```
